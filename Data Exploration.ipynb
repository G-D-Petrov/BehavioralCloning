```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
data_dir = './data/'
data_csv = 'driving_log.csv'
```

```python
training_dat = pd.read_csv(data_dir+data_csv,names=None)
print(training_dat.info())
```

```python
# print a histogram to see which steering angle ranges are most overrepresented
num_bins = 23
avg_samples_per_bin = len(training_dat['steering'])/num_bins
hist, bins = np.histogram(training_dat['steering'], num_bins)
width = 0.7 * (bins[1] - bins[0])
center = (bins[:-1] + bins[1:]) / 2
plt.bar(center, hist, align='center', width=width)
plt.plot((np.min(training_dat['steering']), np.max(training_dat['steering'])), (avg_samples_per_bin, avg_samples_per_bin), 'k-')
plt.show()
```

```python
def normalize_path(img_file):
    img_file = img_file.split('/')[-1]
    img_file = 'data/IMG/'+img_file.split('\\')[-1]
    return img_file
```

```python

for i in range(5):
    left = mpimg.imread(normalize_path(training_dat['left'][i]))
    center = mpimg.imread(normalize_path(training_dat['center'][i]))
    right = mpimg.imread(normalize_path(training_dat['right'][i]))
        
    fig = plt.figure()
    a=fig.add_subplot(i + 2, 3, 1)
    plt.imshow(left)
    a=fig.add_subplot(i + 2, 3, 2)
    plt.imshow(center)
    a=fig.add_subplot(i + 2, 3, 3)
    plt.imshow(right)

    plt.show()
```
